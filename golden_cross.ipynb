# 1. Imports & Ticker List

```python
import yfinance as yf
import pandas as pd
tickers = ['AAPL','NVDA','TSLA','AMD','SMCI','COIN','PLTR','HOOD','MSFT','GOOGL','META','AMZN','NFLX','NOW','CRM','BTC-USD', 'MARA', 'RIOT']
```

# 2. Scanner

```python
golden = []
death = []
for t in tickers:
    try:
        df = yf.download(t, period='2y')[['Close']]
        if len(df) < 200:  # Need 200 days for MA
            continue
        df['50']  = df['Close'].rolling(50).mean()
        df['200'] = df['Close'].rolling(200).mean()
        df = df.dropna()  # Drop NaN rows (first 199 rows)
        df['Prev50']  = df['50'].shift(1)
        df['Prev200'] = df['200'].shift(1)
        df = df.dropna()
        if len(df) < 2:
            continue  # Need today + yesterday
        # Today vs Yesterday
        today_50 = df['50'].iloc[-1]
        today_200 = df['200'].iloc[-1]
        prev_50 = df['Prev50'].iloc[-1]
        prev_200 = df['Prev200'].iloc[-1]
        
        if (today_50 > today_200) and (prev_50 <= prev_200):
            golden.append(t)
        if (today_50 < today_200) and (prev_50 >= prev_200):
            death.append(t)
    except Exception as e:
        print(f"Error {t}: {e}")
```

# 3. Print the Result

```python
print(f"GOLDEN CROSS ({len(golden)}): {', '.join(golden)}")
print(f"DEATH CROSS ({len(death)}): {', '.join(death)}")
```
